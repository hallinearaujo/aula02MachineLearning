```python
 # clustering dataset
 #importando as bibliotecas de machine learning e numpy e matplotlib
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
#definindo o conjunto de dados (manipular)
#criando uma base de dados a partir de dois vetores (objetos np.array) com numeros aleatorios; pega as infos e joga nas variaveis x1 e x2
x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 6, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])
```

```python
#exibindo as informações em formato de grafico para as análises usando o método plotlib

plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset-análise com o k-means')
plt.scatter(x1, x2)
plt.show()
```

```python
# create new plot and data
#cria variavel chamada x que recebe o np.array criando listas separadas por cores (blue, gree, red) e formas (o, v, s)
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']
```

```python
# create new plot and data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'r']
markers = ['o', 'v', 's']

# KMeans algorithm 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

plt.plot()
for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])

plt.show()
```

```python
#o grafico acima esta em cores e formas diferentes para melhor visualização dos agrupamentos
```

```python
#Dataset IrisPython

import pandas as pd #import para manipular os dados
iris = pd.read_csv("iris.csv") #carregar os dados e colocar no data frame chamado Iris
iris.head() #mostrar as 5 primeiras linhas do dataframe
```

```python
#remover a coluna de especie no final do data frame; o comando do pandas iloc[:, 0:4].values -> pega de 0 até as 4 primeiras colunas ignorando a ultima coluna (especie) pois esse é o target
#se quiser pegar colunas especificas, por exemplo -> iloc[:, 0; 5;7;15].values

X = iris.iloc[:, 0:4].values
X
```

```python
#instanciar o objeto do KMeans

kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
#comando fit serve para treinar o modelo de ML com o algoritmo KMeans (não-supervisionado)
kmeans.fit(X)
```

```python
#cluster center exibe a quantidade de centróides (bolinha que fica no meio do cluster, pontos dentrais encontrados entre os dados que verifica a distancia de um grupo para outro) identificados
kmeans.cluster_centers_
```

```python
#exibir a distância entre os pontos centrais identificando todos os pontos que tem dentro do agrupamento
distance = kmeans.fit_transform(X)
distance
```

```python
#comando labels exibe os rótulos dos clusters criando uma quarta coluna, virtual
labels = kmeans.labels_
labels
```

```python
#metodo que indica o valor ideal do "k"; no exemplo acima k=3 pois ja conhecemos a base de dados e sabemos que existe três grupos de flores
#mas num BD muito extenso fica dificil identificar a quantidade de grupos, para esse usamos o Método Elbow
# Método Elbow -> O método Elbow se trata de uma técnica interessante para encontrar o valor ideal do parâmetro k.
# Basicamente o que o método faz é testar a variância dos dados em relação ao número de clusters.

wcss = []
i = 0 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print (i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

```python
# Agrupando Novos Dados com KMeans
# Com o nosso modelo já criado podemos usar o algoritmo K-means para agrupar novos dados baseado nos centroides que já estão definidos.

# Passando novos valores e observando os cluster que serão atribuídos.
```

```python
#Agrupando novos dados KMeansPython
data = [
        [ 4.12, 3.4, 1.6, 0.7],
        [ 5.2, 5.8, 5.2, 6.7],
        [ 3.1, 3.5, 3.3, 3.0]
    ]
kmeans.predict(data)
```

```python
# Visualizando os resultados
# Para ficar ainda mais claro é interessante visualizar através de uma representação gráfica os nossos dados e os centroides criados pelo K-means Veja com o código abaixo como é simples.
```

```python
plt.scatter(X[:, 0], X[:,1], s = 100, c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()
 
plt.show()
```
